```python
# Carica entrambi i file
df_alunni = pd.read_csv(r"C:\Users\pasqu\Desktop\Data Analyst\Progetto scuole\CODE\csv\Alunni 2024.csv")
df_scuole = pd.read_csv(r"C:\Users\pasqu\Desktop\Data Analyst\Progetto scuole\CODE\csv\Anagrafica Scuola 2025-26.csv")

# Verifica match
codici_alunni = set(df_alunni['CODICESCUOLA'].unique())
codici_scuole = set(df_scuole['CODICESCUOLA'].unique())

# DOMANDA CHIAVE: Tutte le scuole in Alunni esistono in Anagrafica?
mancanti = codici_alunni - codici_scuole

print(f"Scuole in Alunni: {len(codici_alunni):,}")
print(f"Scuole in Anagrafica: {len(codici_scuole):,}")
print(f"✓ Scuole presenti in entrambi: {len(codici_alunni & codici_scuole):,}")
print(f"⚠️ Scuole in Alunni MA NON in Anagrafica: {len(mancanti):,}")

if len(mancanti) > 0:
    print("\n⚠️ PROBLEMA! Ci sono studenti senza info scuola")
    print("Esempi codici mancanti:", list(mancanti)[:10])
else:
    print("\n✅ PERFETTO! Tutte le scuole matchano")
```

```python
# Estrai info dalle scuole mancanti
scuole_mancanti_info = df_alunni[df_alunni['CODICESCUOLA'].isin(mancanti)][['CODICESCUOLA']].drop_duplicates()

# Crea dataframe dummy
dummy_scuole = pd.DataFrame({
    'CODICESCUOLA': list(mancanti),
    'DENOMINAZIONESCUOLA': 'Scuola non in anagrafica 2025/26',
    'PROVINCIA': 'NON DISPONIBILE',
    'REGIONE': 'NON DISPONIBILE',
    'AREAGEOGRAFICA': 'NON DISPONIBILE',
    'DESCRIZIONETIPOLOGIAGRADOISTRUZIONESCUOLA': 'NON DISPONIBILE',
    # ... altri campi con valori default
})

# Aggiungi alle scuole esistenti
df_scuole_completo = pd.concat([df_scuole, dummy_scuole], ignore_index=True)
```

```python
# Studenti nelle scuole MANCANTI
studenti_persi = df_alunni[df_alunni['CODICESCUOLA'].isin(mancanti)]['ALUNNI'].sum()

# Studenti TOTALI
studenti_totali = df_alunni['ALUNNI'].sum()

# Studenti che MANTENIAMO
studenti_ok = df_alunni[~df_alunni['CODICESCUOLA'].isin(mancanti)]['ALUNNI'].sum()

# Percentuali
perc_persi = (studenti_persi / studenti_totali) * 100
perc_ok = (studenti_ok / studenti_totali) * 100

# Stampa risultati
print("="*60)
print("ANALISI IMPATTO SCUOLE MANCANTI")
print("="*60)
print(f"📊 Studenti TOTALI: {studenti_totali:,}")
print(f"✅ Studenti in scuole con anagrafica: {studenti_ok:,} ({perc_ok:.2f}%)")
print(f"⚠️ Studenti in scuole SENZA anagrafica: {studenti_persi:,} ({perc_persi:.2f}%)")
print("="*60)

if perc_persi > 5:
    print("⚠️ ATTENZIONE: Perdiamo più del 5% degli studenti!")
    print("→ Consiglio: Creare righe dummy")
elif perc_persi > 1:
    print("⚠️ Perdiamo tra 1-5% degli studenti")
    print("→ Consiglio: Creare righe dummy")
else:
    print("✅ Perdiamo meno dell'1% degli studenti")
    print("→ Opzionale: Possiamo escluderli o creare dummy")
```

```python
# Ricarica/ricontrolla con il nuovo df_scuole_completo
codici_scuole_completo = set(df_scuole_completo['CODICESCUOLA'].unique())
codici_alunni = set(df_alunni['CODICESCUOLA'].unique())

# Adesso dovrebbe essere match perfetto
mancanti_dopo = codici_alunni - codici_scuole_completo

print("="*60)
print("VERIFICA POST-DUMMY")
print("="*60)
print(f"Scuole in Alunni: {len(codici_alunni):,}")
print(f"Scuole in Anagrafica completa: {len(codici_scuole_completo):,}")
print(f"Scuole mancanti: {len(mancanti_dopo)}")
print("="*60)

if len(mancanti_dopo) == 0:
    print("✅✅✅ PERFETTO! Match al 100%!")
else:
    print(f"⚠️ Ancora {len(mancanti_dopo)} scuole mancanti")
```

```python
# Aggiungi colonna flag
df_scuole_completo['SCUOLA_ATTIVA_2025'] = df_scuole_completo['CODICESCUOLA'].apply(
    lambda x: 'NO' if x in mancanti else 'SI'
)
```

```python
# Salva Anagrafica completa (con dummy)
df_scuole_completo.to_csv('anagrafica_scuole_completa.csv', index=False)

# Salva Alunni (già pulito)
df_alunni.to_csv('alunni_cleaned.csv', index=False)

# Hai già personale_cleaned.csv
```
